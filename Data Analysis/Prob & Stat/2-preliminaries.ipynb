### Indentation, loops, and conditional statements

````python
for i in range(10):
    print("Counting to ten: ", i)
````

Note: the loop starts at zero, and ends one before the number we put in. Not many other languages are like this, many will include the end points. If you put 10 in, you will do 10 loops, but we start at 0, and so this happens.

Always remember that the loop will finish at N-1

````python
for ientry in mixedlist:
    print("Band: ", ientry[0], "| Score: ", ientry[1])
````

````python
for i, ientry in enumerate(mixedlist):
    print("Counter: ", i, " | Band: ", ientry[0], " | Score: ", ientry[1])
````

##### Exercise 2.13

````python
for i, ientry in enumerate(sortedlist):
    print("Counter: ", i, " | Band: ", ientry[0], " | Score: ", ientry[1])
````

````python
# while loops

icecreams = ["magnum", "maxibon", "cornetto"]
count = 0
myicecream = "maxibon"
while count < 4:
    print(icecreams[count])
    if icecreams[count] != myicecream:
        count += 1
        continue
    print("I have found my icecream")
    break
````

Flow Controls:

```continue``` jumps to the next iteration of the loop, without finishing the rest of the block

```break``` end the loop right there, and jumps to after the loop without finishing the remaining iterations

````python
a = 1
if a == 1:
    print("a is 1")
elif a == 2:
    print("a is 2")
else:
    print("a is neither 1 or 2")
````

#### List comprehension

````python
sqrnumber = [i * i for i in range(20)]
print(sqrnumber)
````

````python
# only if i is odd

sqrnumber = [i * i for i in range(20) if i % 2 != 0]
print(sqrnumber)
````

````python
# ternary list comprehension

sqrnumber = [i * i if (i % 2 != 0) else "Even" for i in range(20)]
print(sqrnumber)
````

Note the difference in ordering for a conditional (... for ... if ...) vs. a ternary statement (... if ... else ... for ...)

##### Exercise 2.14

````python
cube_even_numbers = [i**3 for i in range(60) if i % 2 == 0]
print(cube_even_numbers)
print(len(cube_even_numbers))
````

Note the need to loop from 0 to 60 (not including the end point), as only half of these numbers are even, so we go double the distance to obtain the first 30 even cubes.

````python
from random import seed, random, randint

seed(1)
for i in range(10):
    value = random()
    print(value)
````

##### Exercise 2.18

We get new numbers, as the random number generator continues from the previous state it was left at.

##### Exercise 2.19

````python
seed(1)
for i in range(10):
    value = random()
    print(value)
````

We get the same numbers we got the first time. The state of the random number generator is reset to the same state it was in before the first generation. This allows us to have consistent results when using random numbers.

````python
min = -2
max = 2
for i in range(10):
    value = random()
    scaled_value = min + (value * (max - min))
    print(scaled_value)
````

````python
from random import gauss

for i in range(10):
    value = gauss(0, 1)
    print(value)
````

### File input and output (I/O)
