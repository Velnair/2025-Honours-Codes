Question 1
==========

Given the linear system of equations with boundary conditions:

$$
ψ_0(x) = A_0e^{ik_0x} + B_0e^{-ik_0x}
\\ψ_1(x) = A_1e^{ik_1(x-x_0)} + B_1e^{-ik_1(x-x_0)}
$$
$$
\\ ψ_0(x_0) = ψ_1(x_0)
\\ \frac{d}{dx}ψ_0(x_0) = \frac{d}{dx}ψ_1(x_0)
$$

When explicitly represented, the boundary conditions show:

$$
A_0e^{ik_0x_0} + B_0e^{-ik_0x_0} = A_1 + B_1
\\ ik_0(A_0e^{ik_0x_0} - B_0e^{-ik_0x_0}) = ik_1(A_1 - B_1)
$$

Equate both boundary conditions to isolate in terms of $A_1$, which gives:

$$
A_1 = A_0e^{ik_0x_0} + B_0e^{-ik_0x_0} - B_1
\\ A_1 = \frac{k_0}{k_1} \left( A_0e^{ik_0x_0} - B_0e^{-ik_0x_0} \right) + B_1
$$

Isolate $B_1$ & solve:

$$
A_0e^{ik_0x_0} + B_0e^{-ik_0x_0} - B_1 = \frac{k_0}{k_1} \left( A_0e^{ik_0x_0} - B_0e^{-ik_0x_0} \right) + B_1
\\ B_1 = \frac{1}{2}\left(A_0e^{ik_0x_0}(1-\frac{k_0}{k_1}) + B_0e^{-ik_0x_0}(1+\frac{k_0}{k_1})\right)
$$

Substite back into either equations of $A_1$ gives:

$$
A_1 = A_0e^{ik_0x_0} + B_0e^{-ik_0x_0} - \frac{1}{2}\left(A_0e^{ik_0x_0}(1-\frac{k_0}{k_1}) + B_0e^{-ik_0x_0}(1+\frac{k_0}{k_1})\right)
\\ A_1 = \frac{1}{2}\left(A_0e^{ik_0x_0}(1+\frac{k_0}{k_1}) + B_0e^{-ik_0x_0}(1-\frac{k_0}{k_1})\right)
$$

This gives the Matrix representation and the sought after $2\times2$ matrix $M_0(E)$, since $k_j(E) = \sqrt{ \frac{2m}{\bar{h}^2} (E-V_j) }, j \in \{ 0,1\}$:

$$
\begin{bmatrix}
A_1
\\ B_1
\end{bmatrix}

= 
\frac{1}{2}
\begin{bmatrix}
e^{ik_0x_0}(1+\frac{k_0}{k_1}) & e^{-ik_0x_0}(1-\frac{k_0}{k_1})
\\ e^{ik_0x_0}(1-\frac{k_0}{k_1}) & e^{-ik_0x_0}(1+\frac{k_0}{k_1})
\end{bmatrix}

\begin{bmatrix}
A_0
\\ B_0
\end{bmatrix}
$$

Therefore, $M_0(E) = \frac{1}{2}
\begin{bmatrix}
e^{ik_0x_0}(1+\frac{k_0}{k_1}) & e^{-ik_0x_0}(1-\frac{k_0}{k_1})
\\ e^{ik_0x_0}(1-\frac{k_0}{k_1}) & e^{-ik_0x_0}(1+\frac{k_0}{k_1})
\end{bmatrix}$.

Question 2
==========

The matrix $M_0(E)$ can be generalised for wavefunctions $ψ_j(x)$ of $j \in \{0,1,2,...\}$, where $ψ_j(x_{j-1}) = A_j + B_j$. Therefore, when absolving boundary conditions, the exponentials for the matrix become:

$$
e^{\pm ik_0x_0} \rightarrow e^{\pm ik_{j-1}(x_{j-1}-x_{j-2})}
$$

The wave constants in the matrix can also be generalised as $k_1/k_0 \implies k_{j}/k_{j-1}$ to transform the constants $\begin{bmatrix}
A_j
\\ B_j
\end{bmatrix} \implies 
\begin{bmatrix}
A_{j-1}
\\ B_{j-1}
\end{bmatrix}$. 

Therefore, for the generalised
$
M_{j-1}(E) = \frac{1}{2}
\begin{bmatrix}
e^{ik_{j-1}(x_{j-1} - x_{j-2})}(1+\frac{k_{j-1}}{k_j}) & e^{-ik_{j-1}(x_{j-1} - x_{j-2})}(1-\frac{k_{j-1}}{k_j})
\\ e^{ik_{j-1}(x_{j-1} - x_{j-2})}(1-\frac{k_{j-1}}{k_j}) & e^{-ik_{j-1}(x_{j-1} - x_{j-2})}(1+\frac{k_{j-1}}{k_j})
\end{bmatrix}
$

Question 3
==========

For the $j^{th}$ layer described by the effective mass $m^*_j$ and potential $V_j$, this would make the propagation constant a function of energy, $E$:

$$
k_j(E, V_j, m^*_j) = \sqrt{ \frac{2m^*_j}{\bar{h}^2} (E-V_j) }
$$

However, the dimensions must be adhered to. In regards to all the parameters in the matrix,the only function that is of dimension is the wavenumber, $k$, in reciprocal length. By dimensional analysis:

$$
k = \sqrt{ \frac{2m^*_j}{\bar{h}^2} (E-V_j) }
\\ = \sqrt{ \frac{[kg]}{[J^2][s^2]} [eV] }
\\ = \sqrt{ \frac{[kg]}{[[eV^2][s^2]} [eV] }
\\ = \sqrt{ \frac{[eV]/[c^2]}{[eV][s^2]} }
\\ = \frac{1}{[c][s]}
\\ = \frac{[s]}{[m][s]}
\\ = \frac{1}{[m]}
$$

Since the wavenumber is in units of reciprocal length, the units of mass must be chosen accordingly. Constants & variables must be in these dimensions, $\bar{h} = [eV][s]$, $E = [eV]$, $m = [eV][s^2]/[m^2]$, $L = [m]$ as the length must match the dimensions of reciprocal length in order to be dimensionless in the exponentials.

Question 4
==========

For the $j-1^{th}$ layer width $L_{j-1}$, this describes that $x_{j-2} = x_{j-1} + L_{j-1}$. Rearranging rewrites the exponentials in the matrix to:
$$
e^{\pm ik_{j-1}(x_{j-1}-x_{j-2})} \rightarrow e^{\pm ik_{j-1}L_{j-1}}
$$
Since the layers of different materials have different effective masses, the continuity boundary conditions change to

$$
\\ \frac{1}{m^*_{j-1}} \frac{d}{dx}ψ_{j-1}(x_{j-1}) = \frac{1}{m^*_j} \frac{d}{dx}ψ_j(x_{j-1})
$$

Since this condition is affected just by a factor, this gives rise to the constant $\alpha = \frac{m^*_j}{m^*_{j-1}}\frac{k_{j-1}}{k_j}$ within the matrix, since the arthimatic follows through the same steps as in Question 1. Note that due to the layer width in the exponentials, the matrix itself becomes dimensionless as well. In addition to the wave constant as a function of the layer characteristics, gives the overall matrix function:

$$
M_{j-1}(E,L_{j-1}, [V_j, V_{j-1}], [m^*_j,m^*_{j-1}]) = \frac{1}{2}
\begin{bmatrix}
e^{ik_{j-1}L_{j-1}}(1+\alpha) & e^{-ik_{j-1}L_{j-1}}(1-\alpha)
\\ e^{ik_{j-1}L_{j-1}}(1-\alpha) & e^{-ik_{j-1}L_{j-1}}(1+\alpha)
\end{bmatrix}
$$

Question 5
==========

The bandgap for this 1D Quantum Well would be $\Delta E = 1.1x$, where $x = 30\%$ doping of the material $AlGaAs$. However, since the dopant percentage is less than $45\%$, then the overall bandgap would be $60\%$ value calculated of the bandgap energy. This gives out to $\Delta E = 198 meV$, the depth of the potential well. Now given the variables $E,[0, 20nm], [0, 198meV], [m^*_{AlGaAs},m^*_{GaAs}]$, the transfer matrix can be obtained by simple matrix multiplication. The initial matrix $M_0$ describes the wavefunction at $x_0 = 0nm $ with $V = 0meV$, transitioning from solids $m_{AlGaAs} \rightarrow m_{GaAs}$. The final matrix, $M_1$, located at $x_1 = 20nm$ would have the energy bandgap $V = 198meV$, and the materials inverted $m_{GaAs} \rightarrow m_{AlGaAs}$. This gives the transfer function for this 1D well:
$$
TF = M_1M_0
$$

```python
import numpy as np
from numpy import exp, cos, sin, pi, sqrt
import matplotlib.pyplot as plt
import scipy as sp

# constants, in base eV
meV = 1e3  # meV scaling
h = 6.582e-16 # eV, hbar
L = 20   # nm, assume quantum well is symmetric
x = 0.3  # doped at 30%

# Unsure as to why the functions behave well when the rest mass is in [kg], when dimensions expected to be in [eV]/[s]^2
m0 = 9.109e-31 # kg, this should be in ev/c^2 for dimension of the k-number?
#m0 = 0.511e6*meV # eV/c^2 -> meV/c^2, factor of 1e6 for MeV/c^2

V = [0j, (1.1*x*0.6)*meV + 0j] # meV, e is outside well when at bandgap energy, where bandgap is 60% from difference if x < 0.45
points = 100
E = np.linspace(5,150,points, dtype = 'float128')    # meV units
m_GaAs, m_AlGaAs = 0.067*m0, (0.067 + 0.083*(0.3))*m0    # meV/c^2
print(V)

def k(E, V, m):    # [k] = 1/m
    return sqrt(2*m/(h**2)/160.2 * (E - V) )  

def α(E,V_f,V_i,m_f,m_i):    # dimensionless, {j-1,j} = {initial, final}
    return m_f/m_i *  k(E,V_i,m_i) / k(E,V_f,m_f)  

def M(E,L,V_f,V_i,m_f,m_i):    # dimensionless
    return 1/2*np.array([ [exp(1j*k(E,V_i,m_i)*L)*(1 + α(E,V_f,V_i,m_f,m_i)), exp(-1j*k(E,V_i,m_i)*L)*(1 - α(E,V_f,V_i,m_f,m_i))],
                          [exp(1j*k(E,V_i,m_i)*L)*(1 - α(E,V_f,V_i,m_f,m_i)), exp(-1j*k(E,V_i,m_i)*L)*(1 + α(E,V_f,V_i,m_f,m_i))] ])

def TF(E, L, L_b, normalise = False):
    
    T22_abs = np.empty(points)
    T21_abs = T22_abs.copy()
    T12_abs = T22_abs.copy()
    T11_abs = T22_abs.copy()


    
    for idx, energy in enumerate(E):         
        # from AlGaAs -> GaAs
        # V_final,V_initial = 0,1
        # m_final,m_initial = m_GaAs, m_AlGaAs
        M0 = M(energy,L_b,V[0],V[1],m_GaAs,m_AlGaAs)
        alpha_0 = α(energy,V[0],V[1],m_GaAs,m_AlGaAs)
    
        # from GaAs -> AlGaAs
        # V_final, V_initial = 1,0
        # m_final,m_initial = m_AlGaAs,m_GaAs
        M1 = M(energy,L,V[1],V[0],m_AlGaAs,m_GaAs)
        alpha_1 = α(energy,V[1],V[0],m_AlGaAs,m_GaAs)
        
        if normalise == True:

            # for T12 = A_2/B_0,
            # M1[0,0] = (exp(1j*k(energy,V[1],m_GaAs)*L)*M0[0,1] + exp(-1j*k(energy,V[1],m_GaAs)*L)*M0[1,1] - M1[0,1]*M0[1,1])/M0[0,1], 
            
            M1[1,1] = M1[1,0] # for T12 = A_2/B_0

            # for T22 = 0,
            #M1[1,0] = M1[1,1]*M0[1,1]/M0[0,1]
 
            beta = sqrt((2 + alpha_1 + alpha_0)/(2 - alpha_1 - alpha_0))
            M1[0,0] = beta*(1+alpha_1) 
            M1[1,0] = beta*(1-alpha_1)
            
            
             
        TF = M1@M0
    
        T11_abs[idx] = abs(TF[0,0])**2
        T12_abs[idx] = abs(TF[0,1])**2
        T21_abs[idx] = abs(TF[1,0])**2
        T22_abs[idx] = abs(TF[1,1])**2
   
    return T11_abs, T12_abs, T21_abs, T22_abs
    

T11_abs, T12_abs, T21_abs, T22_abs = TF(E,20,0)

plt.plot(E,T11_abs, label = '$|T_{11}|$')
plt.plot(E,T12_abs, label = '$|T_{12}|$')
plt.plot(E,T21_abs, ls = '--', label = '$|T_{21}|$')
plt.plot(E,T22_abs, ls = '--', label = '$|T_{22}|$')
plt.xlabel('E, [meV]')
plt.ylabel('Power Spectrum')

plt.legend()
plt.grid()
```

Question 6
==========

The transmittence & reflectance of the wavefunction is given by:

$$
r = -\frac{T^{21}}{T^{22}}
\\ t = \frac{T^{11}T^{22}-T^{12}T^{21}}{T^{22}}
\\
\\ |t|^2 + |r|^2 = 1
$$

The functions must first be normalised.The conditions must be met for the final transfer matrix:

$$
TF = \begin{bmatrix}
T^{11} & T^{12}
\\ T^{21} & T^{22}
\end{bmatrix}
= 
\begin{bmatrix}
M_1^{11} & M_1^{12}
\\ M_1^{21} & M_1^{22}
\end{bmatrix}

\begin{bmatrix}
M_0^{11} & M_0^{12}
\\ M_0^{21} & M_0^{22}
\end{bmatrix}
\\
\\T^{22} = 0
\\T^{12} = A_2/B_0
$$

Since the Transfer Matrix is a product of two matrices, the entries of the Transfer Matrix is a product of the previous matrices to describe wavefunction transformations. Hence:

$$
T^{22} = M_1^{21}M_0^{12} + M_1^{22}M_0^{22} = 0
\\T^{12} = M_1^{11}M_0^{12} + M_1^{12}M_0^{22} = A_2/B_0
$$

For solving the condition of $T^{12}$, for shorthand consider the exponentials of their respective matrices denoted by $e_{0,1}^\pm$. Also note that $\alpha_0 \alpha_1 = 1$, since the quantum well involves transition between two materials only. Recall that for all constants is dependent on the initial constant, $A_2(A_0,B_0)$:
$$
A_2(A_0,B_0) = A_1(A_0,B_0)e_1^+ + B_1(A_0,B_0)e_1^-
\\ = (A_0 M_0^{11} + B_0 M_0^{12})e_1^+ + (A_0 M_0^{21} + B_0 M_0^{22})e_1^-
\\ \implies A_2/B_0 = (A_0 M_0^{11} + B_0 M_0^{12})e_1^+ + (A_0 M_0^{21} + B_0 M_0^{22})e_1^-/B_0
$$

But since $A_0 = B_2 = 0$ for bound states to be normalisable, the $A_0$ terms fall to zero. This gives the equation to solve for the condition of $T^{12}$:

$$
M_1^{11}M_0^{12} + M_1^{12}M_0^{22} = e_1^+M_0^{12} + e_1^-M_0^{22}
\\ [e_1^+(1+\alpha_1)] [e_0^-(1-\alpha_0)] + [e_1^-(1-\alpha_1)] [e_0^-(1+\alpha_0)] = e_1^+ [e_0^-(1-\alpha_0)] + e_1^- [e_0^-(1+\alpha_0)]
\\ e_1^+(\alpha_1 - \alpha_0) + e_1^-(\alpha_0 - \alpha_1) = e_1^+(1-\alpha_0) + e_1^-(1+\alpha_0)
\\ e_1^+(1-\alpha_1) = e_1^-(1+\alpha_1)
\\ \therefore M_1^{21} = M_1^{22}
$$

Similarly, second normalisation condition can be found for $T^{22}$:

$$
M_1^{21}M_0^{12} = M_1^{22}M_0^{22}
\\ [e_1^+(1-\alpha_1)][e_0^-(1-\alpha_0)] = [e_1^-(1+\alpha_0)][e_0^-(1+\alpha_0)]
\\ e_1^+(2-\alpha_0-\alpha_1) = e_1^-(2+\alpha_0+\alpha_1)
\\ \therefore e_1^+ = \sqrt{ \frac{(2+\alpha_0+\alpha_1)}{(2-\alpha_0-\alpha_1)} } = \beta
$$

Since both of the conditions affect the latter transformation matrix, the condition gives the new matrix which must be matrix multiplied by the initial matrix $M_0$ to obtain a normalised transfer matrix of the 1D Quantum Well:

$$
M_1 = 
\begin{bmatrix}
\beta(1+\alpha_1) & e^{-ik_1L}(1-\alpha_1)
\\ \beta(1-\alpha_1) & \beta(1-\alpha_1)
\end{bmatrix}
$$

Since initially the barrier width was set to $0nm$, this width would be altered for the following. The well width would still stay $20nm$, since this would be describing scattering of the incident wavefunction on the materials.

```python
# create function to obtain reflectance & transmittance
def tran_ref(T11_abs,T12_abs,T21_abs,T22_abs):
    r = -sqrt(T21_abs)/sqrt(T22_abs)
    t = (sqrt(T11_abs)*sqrt(T22_abs)-sqrt(T12_abs)*sqrt(T21_abs))/sqrt(T22_abs)
    return r,t

r,t = tran_ref(*TF(E,20,1,True))    # barrier width 1nm

# unfortunately with troubleshooting I cannot get it to normalise
print('|r|^2 + |t|^2 = 1, ',np.all(abs(t)**2 +  abs(r)**2 == 1))

plt.figure()
plt.plot(E,abs(t)**2, label = '1nm')

r,t = tran_ref(*TF(E,20,6,True))    # barrier width 6nm

plt.plot(E,abs(t)**2, label = '6nm')

r,t = tran_ref(*TF(E,20,10,True))   # barrier width 10nm

plt.plot(E,abs(t)**2, label = '10nm')
plt.xlabel('E, [meV]')
plt.ylabel('Transmittance')
plt.grid()
plt.legend()
```
