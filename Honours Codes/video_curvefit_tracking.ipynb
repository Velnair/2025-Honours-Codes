```python
import numpy as np
from scipy.special import assoc_laguerre
import matplotlib.pyplot as plt

def laguerre_gaussian_beam(x, y, A, x0, y0, p, l, w_z):
    """
    Generates the complex amplitude of a Laguerre-Gaussian beam.

    Args:
        x (numpy.ndarray): Array of x-coordinates.
        y (numpy.ndarray): Array of y-coordinates.
        x0: offset centering in x-coordinates
        y0: offset centering in y-coordinates
        p (int): Radial mode number.
        l (int): Azimuthal mode number (topological charge).
        w0 (float): Beam waist radius at z=0.
        z (float, optional): Propagation distance along the z-axis. Defaults to 0.
        k (float, optional): Wave number (2*pi/lambda). Defaults to 1.
        zR (float, optional): Rayleigh range. If None, it's calculated from w0 and k.

    Returns:
        numpy.ndarray: Complex amplitude of the Laguerre-Gaussian beam.
    """
    # if zR is None:
    #     zR = np.pi * w0**2 * k  # Assuming k = 2*pi/lambda for a given lambda

    r_sq = (x-x0)**2 + (y-y0)**2
    #r = np.sqrt(r_sq)
    # phi = np.arctan2(y, x)

    # w_z = w0 * np.sqrt(1 + (z / zR)**2)
    # R_z = z * (1 + (zR / z)**2) if z != 0 else np.inf
    # gouy_phase = (2 * p + np.abs(l) + 1) * np.arctan(z / zR)

    # Laguerre polynomial term
    laguerre_poly = assoc_laguerre(2 * r_sq / w_z**2, p, np.abs(l))

    # Complex amplitude
    amplitude = A*np.exp(-r_sq / w_z**2) * laguerre_poly

    return amplitude

# Example usage:
if __name__ == "__main__":
    # Beam parameters
    l_mode = 1  # Azimuthal mode number
    p_mode = 0  # Radial mode number
    beam_waist = 1.0  # Beam waist radius
    wavelength = 0.632e-6  # Wavelength (e.g., HeNe laser)
    k_wavenumber = 2 * np.pi / wavelength

    # Spatial grid
    grid_size = 100
    x_coords = np.linspace(-5 * beam_waist, 5 * beam_waist, grid_size)
    y_coords = np.linspace(-5 * beam_waist, 5 * beam_waist, grid_size)
    X, Y = np.meshgrid(x_coords, y_coords)

    # Generate LG beam
    lg_beam = laguerre_gaussian_beam(X, Y, 1, -1, p_mode, l_mode, beam_waist, k=k_wavenumber)

    # Calculate intensity and phase
    intensity = np.abs(lg_beam)**2
    phase = np.angle(lg_beam)

    # Plotting
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(intensity, extent=(x_coords.min(), x_coords.max(), y_coords.min(), y_coords.max()),
               cmap='hot', origin='lower')
    plt.colorbar(label='Intensity')
    plt.title(f'Intensity Profile (l={l_mode}, p={p_mode})')
    plt.xlabel('x')
    plt.ylabel('y')

    plt.subplot(1, 2, 2)
    plt.imshow(phase, extent=(x_coords.min(), x_coords.max(), y_coords.min(), y_coords.max()),
               cmap='hsv', origin='lower')
    plt.colorbar(label='Phase (rad)')
    plt.title(f'Phase Profile (l={l_mode}, p={p_mode})')
    plt.xlabel('x')
    plt.ylabel('y')

    plt.tight_layout()
    plt.show()
    
```

```python
import numpy as np
from numpy import exp
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.ndimage import gaussian_filter
import cv2

def laguerre_gaussian_beam(x, y, x0, y0, p, l, w0, z=0, k=1, zR=None):
    """
    Generates the complex amplitude of a Laguerre-Gaussian beam.

    Args:
        x (numpy.ndarray): Array of x-coordinates.
        y (numpy.ndarray): Array of y-coordinates.
        x0: offset centering in x-coordinates
        y0: offset centering in y-coordinates
        p (int): Radial mode number.
        l (int): Azimuthal mode number (topological charge).
        w0 (float): Beam waist radius at z=0.
        z (float, optional): Propagation distance along the z-axis. Defaults to 0.
        k (float, optional): Wave number (2*pi/lambda). Defaults to 1.
        zR (float, optional): Rayleigh range. If None, it's calculated from w0 and k.

    Returns:
        numpy.ndarray: Complex amplitude of the Laguerre-Gaussian beam.
    """
    if zR is None:
        zR = np.pi * w0**2 * k  # Assuming k = 2*pi/lambda for a given lambda

    r_sq = (x-x0)**2 + (y-y0)**2
    r = np.sqrt(r_sq)
    phi = np.arctan2(y, x)

    w_z = w0 * np.sqrt(1 + (z / zR)**2)
    R_z = z * (1 + (zR / z)**2) if z != 0 else np.inf
    gouy_phase = (2 * p + np.abs(l) + 1) * np.arctan(z / zR)

    # Laguerre polynomial term
    laguerre_poly = assoc_laguerre(2 * r_sq / w_z**2, p, np.abs(l))

    # Complex amplitude
    amplitude = (w0 / w_z) * (r * np.sqrt(2) / w_z)**np.abs(l) * \
                np.exp(-r_sq / w_z**2) * laguerre_poly * \
                np.exp(-1j * (k * r_sq / (2 * R_z) + l * phi - gouy_phase))

    return amplitude

def LG_ravel(t, *p):
    """
    t = [x,y]       [3D object]
    *p = parameters [array] 
    """
    x,y = t
    return laguerre_gaussian_beam(x, y, *p).ravel()

```

```python
# camera characteristics 
w,h = 720, 540
guess = [160, w//2,50,h//2,50]
w,h = np.meshgrid(w,h)
fps = 33
duration = 31
frames = fps*duration

# fitting function & ravelling
def gauss_2D(x, y, A, x_mean, x_varience, y_mean, y_varience):
    return ( A*exp(-1/2*(x-x_mean)**2/x_varience)
                *exp(-1/2*(y-y_mean)**2/y_varience) )

def gauss_ravelled(t, *p):
    # *parameters here is used to forward on all position arguments to the
    # gaussian_2D, it saves rewriting the arguments out
    
    # unpack t back into x and y
    x = t[0] # is this a list?
    y = t[1]
    # Now we flatten (ravel) the model
    return gauss_2D(x, y, *p).ravel()

t = np.array([w,h])
fitted = []

# --- Load the video ---
video_path = "FLIR_translational_GrayScale.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    raise FileNotFoundError(f"Could not open video: {video_path}")

frame_idx = 0

while True:
    ret, frame = cap.read()   # Read next frame
    if not ret:
        break                 # End of video

    #%% [TRACKING ALGORITHM USING CURVEFITTING PARAMETERS]
    z = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # apply Gaussian filter
    z = gaussian_filter(z, sigma=10)
    
    try:
        p, cov = curve_fit(gauss_ravelled, t, z.ravel(), p0 = guess)
        fitted.append(p)     # store the vector
        guess = p                # recursive update
        plt.figure()
        plt.scatter(p[1],p[3])
        plt.savefig(f'{frame_idx}')
        
    except RuntimeError:
        print(f"Fit failed on frame {frame_idx}")
        fitted.append([np.nan]*len(guess))  # placeholder if fit fails

    frame_idx += frame_idx
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
    print(f'{int(frame_idx/frames*100)}% Processed...', end = '\r')

print('Done!')

cap.release()
cv2.destroyAllWindows()

#%% [PLOT THE COORDINATE CENTERS OF THE GAUSSIAN]
fitted = np.array(fitted)
plt.scatter(fitted[1,:],fitted[3,:])
```

```python
# --- Camera characteristics ---
W, H = 720, 540
x, y = np.meshgrid(np.arange(W), np.arange(H))
t = np.array([x, y])

# --- Initial guess: A, x_mean, x_var, y_mean, y_var
guess = [1, W//2, 50, H//2, 50]

# --- Gaussian model ---
def gauss_2D(x, y, A, x_mean, x_var, y_mean, y_var):
    return (A * np.exp(-0.5 * ((x - x_mean)**2 / x_var)
                       -0.5 * ((y - y_mean)**2 / y_var)))

def gauss_ravelled(t, *p):
    x, y = t
    return gauss_2D(x, y, *p).ravel()

# --- Load the video ---
video_path = "FLIR_translational_GrayScale.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    raise FileNotFoundError(f"Could not open video: {video_path}")

# --- Storage for fitted parameters ---
fit_params = []

frame_idx = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray - gaussian_filter(gray,sigma=10)

    try:
        p, cov = curve_fit(gauss_ravelled, t, gray.ravel(), p0=guess)
        fit_params.append(p)     # store the vector
        guess = p                # recursive update
        plt.figure()
        plt.xlim((0,W))
        plt.ylim((0,H))
        plt.scatter(p[1,:],p[3,:])
        plt.savefig('Tracked')
    except RuntimeError:
        print(f"Fit failed on frame {frame_idx}")
        fit_params.append([np.nan]*len(guess))  # placeholder if fit fails

    frame_idx += 1
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

fit_params = np.array(fit_params)
np.savetxt("fit_results.txt", fit_params, header="A x_mean x_var y_mean y_var")
print("Saved fit parameters for", len(fit_params), "frames.")
```

```python
# Organize all .txt files into one

import numpy as np

N = 1001
freq = (0.5, 1.0, 5.0, 10.0, 25.0, 50.0, 75.0, 100.0)
V = (0.2,0.4,0.6,0.8)

V_collect = np.zeros((N,len(freq))) # datapoints x frequencies

for v in range(len(V)):
    for i, f in enumerate(freq):
        data = np.loadtxt(f'{V[v]}V-data/{f}Hz-data')   # include extension if needed
        V_collect[:,i] = data
        np.savetxt(f'{V[v]}V', V_collect)
```

```python
# extract each data & obtain mean, standarddeviations and display on

import matplotlib.pyplot as plt
from scipy.stats import *

fig,ax = plt.subplots(1,1,figsize = (20,12))
sigma = np.zeros((len(V),len(freq)))  

for i, v in enumerate(V):
    data = np.loadtxt(f'{v}V')                                                  # extract the data
    mean = data.mean(axis=0)
    std = data.mean(axis=0)
    ax.loglog(freq,mean, marker = 'o', label = f'{v}V')            # data.mean(axis=0) = mean along the rows, gives mean per coloumn. Should have len(V) datapoints
    sigma[i,:] = std.T
    # annotate each datapoint
    # for j in range(data.shape[1]):
    #     ax.text(freq[j],mean[j] + 0.2, f'{std[j]:.2e}')

np.savetxt('sigma', sigma)
ax.legend(fontsize = 26)
ax.grid(True, which="both", ls="-", lw=0.5, color='gray')
ax.set_xlabel('Frequency, [Hz]', fontsize = 16)
ax.set_ylabel('Average Events per Period, [{Evs}/s]', fontsize = 16)
fig.savefig('Events per Period')
```
