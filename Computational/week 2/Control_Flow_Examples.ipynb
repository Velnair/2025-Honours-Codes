```python
# https://docs.python.org/3/tutorial/controlflow.html
import numpy as np
# x_v = np.empty([1,1])
x_v = [] # Defines an empty array
for ii in range(0, 3): # Note the colon at the end and use indentation to complete
    print(ii)
    x_v = np.insert(x_v, ii ,ii**2) # Remember that growing arrays is generally a bad idea for speed
    print(x_v)
```

```python
material = 'Fused Silica'
if material == 'Fused Silica':
    alpha = 0.55e-6
    density = 2.2e3
    specific_heat = 670

elif material == 'Sapphire':
    alpha = 8.5
    density = 3.98e3
    specific_heat = 763
    
elif material == 'Silicon':
    alpha = 4.68
    density = 2.32e3
    specific_heat = 702
else:
    print('Warning incorrectly specified material')

print('Thermal Expansion is ', alpha, '/k'  )
print('Density is', density, 'k m^-3')
print('Specfic heat is', specific_heat,'J/Kg/K')
```
