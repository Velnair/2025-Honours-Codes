```python
# https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.logic.html
import numpy as np
a = 1
b = 2
c = 1
d = 1 +1j
print('a > b',a > b)
print('a ge b', a>=b)
print('a not equal b', a!=b)
print('a equal c', a==c) #Note the double equal sign if querying something equal to something
print('Is d complex', np.iscomplex(d))
```

```python
# Logical Operators
print('Logical Or', np.logical_or(True, False))
print('Logical And', np.logical_and(True,False))
print('Logical XOR', np.logical_xor(True,False))
print('Logical Not', np.logical_not(True))
print('All True', np.all([True,False,True]))
print('Any True', np.any([True,False,True]))
```
