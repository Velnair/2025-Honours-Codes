```python
# Random number generators in Python Numpy
# https://docs.scipy.org/doc/numpy-1.14.0/reference/routines.random.html
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html

import matplotlib.pyplot as plt
import numpy as np
n_points = 1000000 #int(1e6) #if a size is to be defined must be an integer
n_bins = 50
data_flat = np.random.rand(n_points)
data_gauss = np.random.randn(n_points)

plt.hist(data_gauss,n_bins)
plt.savefig('Histogram_gauss.png')

plt.figure(2)
plt.hist(data_flat,n_bins)
plt.savefig('Histogram_Flat.png')
```
