```python
# A Simple plotting exercise in Python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10*np.pi,100)
y1 = np.exp(-0.1*x)*np.sin(x)
y1lim = np.exp(-0.1*x)
plt.figure(1)
plt.plot(x,y1,label = 'Damped Sinusoid')
plt.plot(x,y1lim, label = 'Upper Limit')
plt.plot(x,-y1lim, label = 'Lower Limit')
plt.xlabel('Time(s)')
plt.ylabel('Voltage (Volts)')
plt.title('Plot of Damped Sinusoid')
plt.legend()
```

```python
# A loglog plot with subplots in Python
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.grid.html
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html

w0 = 1 # Corner Frequency
w = np.logspace(-2,2,1000)
tf_v = 1/(1+1j*w/w0)

y1lim = np.exp(-0.1*x)

plt.subplot(211)
plt.loglog(w,np.absolute(tf_v))
plt.grid()
plt.ylabel('Amplitude (V/V)')
plt.subplot(212)
plt.grid()
plt.semilogx(w,np.angle(tf_v))
plt.ylabel('Phase (rad)')
plt.xlabel('w(rad/s)')

```
