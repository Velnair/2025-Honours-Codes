```python
import numpy as np
a = np.array([1,2,3,4]) # 1D array
b = np.array([[1,2,3,4]]) # 2D row vector
c = np.array([[1],[2],[3],[4]]) # 2D col vector
print('a[1]=',a[1])
print(len(a))
print('b[0,1]=', b[0,1])
print(len(b))
print(np.shape(b))
print('c[1,0]=', c[1,0])
print(len(c))
print(np.shape(c))

d = np.transpose(b)
print('np.shape(d)=',np.shape(d))
```
