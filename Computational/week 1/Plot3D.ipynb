```python
# https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Allows 3D plotting
# Simple example of using meshgrid
x_v = np.linspace(-np.pi*6,np.pi*6,100)
y_v = np.linspace(-np.pi*6,np.pi*6,101)
x_m, y_m = np.meshgrid(x_v,y_v)

# Now generate a sinc function
f = np.sin(np.sqrt(x_m**2+y_m**2))/np.sqrt(x_m**2+y_m**2)

# Now plot
fig = plt.figure(1)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_m, y_m, f)
```

```python
# Now create a contour plot
fig = plt.figure(2)
ax = fig.add_subplot(111)
# Now label the contours
CS = ax.contour(x_m, y_m, f) 
ax.clabel(CS, inline=1, fontsize=10)
```

```python
# Now create a wiremesh plot
fig = plt.figure(2)
ax = fig.add_subplot(111,projection='3d')
ax.plot_wireframe(x_m, y_m, f)
```

```python
# Now create a  pcolor plot
fig = plt.figure(3)
ax = fig.add_subplot(111)
CB = ax.pcolor(x_m, y_m, f)
fig.colorbar(CB, ax=ax)
```
