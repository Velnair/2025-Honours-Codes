```python
# Numerical Integration for data
# https://docs.scipy.org/doc/numpy/reference/generated/numpy.trapz.html
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,6*np.pi,100)
ysin = np.sin(x)

# Now integrate data using trapz

ysin_int = np.trapz(ysin,x)
print(ysin_int)
```

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.cumtrapz.html
from scipy import integrate
ysin_int_cum = integrate.cumulative_trapezoid(ysin, x, initial=0)
plt.plot(x,ysin,label = 'Sin(x)')
plt.plot(x,ysin_int_cum, label = 'Integrated Sin(x)')
plt.legend()
```
