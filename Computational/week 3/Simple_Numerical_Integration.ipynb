```python
# Numerical integration of functions
import scipy.integrate as integrate
from numpy import sqrt, sin, cos, pi
def integrand(x):
    integ = sqrt(sin(x**2))
    return integ

res = integrate.quad(integrand,0,sqrt(pi))
print(res)
```

```python
from scipy.integrate import quad
def integrand2(x, a, b):
     return a*x + b
res2 = quad(integrand2,0,1,args=(1,1))
# Note args in this case is the parameters of the functions
print(res2)
```
