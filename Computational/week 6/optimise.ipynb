```python
import scipy.optimize as opt
import numpy as np
import matplotlib.pyplot as plt


def test_function(x):
    return (x-1)**2 - 2


# Note the package root_scalar is not included with anaconda standard
sol = opt.newton(test_function, x0=-1.5)
print("The closest root of the function is", sol)
```

```python
# Now find the minimum
# Scalar means a function that returns a single variable, in this case
sol2 = opt.minimize_scalar(test_function)
print("The minimum of the function is ", sol2.fun)
print("The whole output of minimize_scalar is \n", sol2)
```

```python
# Generate x values for plotting
x = np.linspace(-2, 2, 500)
y = test_function(x)

# Plot the function
plt.plot(x, y, label="f(x) = x^2 - 2")

# Plot the root
plt.scatter(sol, test_function(sol), color='red', label="Root", zorder=5)

# Plot the minimum
plt.scatter(sol2.x, sol2.fun, color='green', label="Minimum", zorder=5)

# Add labels and legend
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Root and Minimum of the Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
